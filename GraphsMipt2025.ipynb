```python
class Queue:
    def __init__(self):
        self.queue = []

    def add(self, data):
        self.queue.append(data)

    def get(self):
        if len(self.queue) > 0:
            return self.queue.pop(0)
        return None

    def is_empty(self):
        return len(self.queue) == 0

def bfs(graph, vertex):
    def bfs_rec(graph, vertex, visited, queue):
        if vertex not in visited:
            print(vertex)
            
            for i, elem in enumerate(graph[vertex]):
                if elem == 1 and i not in visited and i not in queue.queue:
                    queue.add(i)
            visited.append(vertex)
        if not queue.is_empty():
            bfs_rec(graph, queue.get(), visited, queue)

    visited = []
    queue = Queue()
    bfs_rec(graph, vertex, visited, queue)
    return visited
```

```python
class Stack:
    def __init__(self):
        self.queue = []

    def add(self, data):
        self.queue.append(data)

    def get(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        return None

    def is_empty(self):
        return len(self.queue) == 0

def dfs(graph, vertex):
    def dfs_rec(graph, vertex, visited, queue):
        if vertex not in visited:
            print(vertex)
            
            for i, elem in enumerate(graph[vertex]):
                if elem == 1 and i not in visited:
                    queue.add(i)
            visited.append(vertex)
        if not queue.is_empty():
            dfs_rec(graph, queue.get(), visited, queue)

    visited = []
    queue = Stack()
    dfs_rec(graph, vertex, visited, queue)
    return visited
```

```python
def prim(graph):
    spanning = [[math.inf]*len(graph) for _ in range(len(graph))]
    verteces = [0]
    while len(verteces) != len(graph):
        min_edge = math.inf
        min_vertex = None
        for vertex in verteces:
            for i, elem in enumerate(graph[vertex]):
                if elem < min_edge and i not in verteces:
                    min_edge = elem
                    min_vertex = (vertex, i)
        spanning[min_vertex[0]][min_vertex[1]] = min_edge
        spanning[min_vertex[1]][min_vertex[0]] = min_edge
        verteces.append(min_vertex[1])
    return spanning
```
